```python
# I`ve use here for loop for label encoding insted of above 
label_encoder = LabelEncoder()

# Defining the columns to encode 
columns_to_encode = df[["Education", "EmploymentType", "MaritalStatus",
                      "HasMortgage", "HasDependents", "LoanPurpose", "HasCoSigner"]]

# Loop through the columns and encode the columns
for i in columns_to_encode:
    df[i]=label_encoder.fit_transform(df[i])

#Display the updated info 
df
```

```python
df.drop(["LoanID"],axis=1,inplace=True)
```

```python
# we see the correlation after encoding 
plt.figure(figsize=(30,30))
corr = df.corr()
sns.heatmap(corr, vmin = -0.25, annot = True, vmax = 0.6)
plt.title('Correlation Heatmap');
```

```python
# check the output column value difference 
# To check our dataset is balanced or imbalanced. if it is imbalance we have to perform operation called
# OverSampling or UnderSampling
df["Default"].value_counts()
```

```python
# From above output we can see the our Dataset is imbalanced we have to balance it by using OverSampling or UnderSampling
X=df
y=df["Default"]
```

```python
X.drop(["Default"],axis=1,inplace=True)
```

```python
# Apply random oversampling
over_sampler = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = over_sampler.fit_resample(X, y)
```

```python
X_resampled.shape , y_resampled.shape
```

```python
# Our Dataset is balanced now we can perform further operations
```

```python
X_resampled
```

```python
y_resampled.value_counts()
```

```python
# Now to perform feature engineering we have to split data in training and testing 
x_train, x_test, y_train, y_test = train_test_split(X_resampled,y_resampled,test_size=0.2,random_state =53)
```

```python
print('x_train:',x_train.shape)
print('y_train:',y_train.shape)
print('x_test:',x_test.shape)
print('y_test:',y_test.shape)
```

# Convert Dataset into Standard Scaler

```python
# Importing Standard Scaler to convert Data Into Gaussian form
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)

```

# Model Training

```python
# Our First Model is Random Forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
# Train the classifier on the training set
model = model.fit(x_train,y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)
```

```python
# Our Second Model is naive Bayes
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_clf = GaussianNB()
```

```python
# Train the classifier on the training set
model = nb_clf.fit(x_train,y_train)
# Make predictions on the test set
y_pred1 = nb_clf.predict(x_test)
```

```python
# Our third Model is Decision Tree Classifier 
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
# Train the classifier on the training set
model = dt.fit(x_train,y_train)
```

```python
# Make predictions on the test set
y_pred2 = dt.predict(x_test)
```

```python
# Train the classifier on the training set
model = Lr.fit(x_train,y_train)
```

```python
# Make predictions on the test set
y_pred3 = Lr.predict(x_test)
```

# Model Evaluation

```python
# Importing the evaluation metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Evaluate the model Random Forest Classifier 
accuracy_RF = accuracy_score(y_test, y_pred)
conf_matrix_RF = confusion_matrix(y_test, y_pred)
classification_rep_RF = classification_report(y_test, y_pred)

# Print the evaluation metrics
print("Accuracy:", accuracy_RF)
print("\nConfusion Matrix:")
print(conf_matrix_RF)
print("\nClassification Report:")
print(classification_rep_RF)
```

```python
# Evaluate the model Naive Bayes Classifier 
accuracy_NB = accuracy_score(y_test, y_pred1)
conf_matrix_NB = confusion_matrix(y_test, y_pred1)
classification_rep_NB = classification_report(y_test, y_pred1)

# Print the evaluation metrics
print("Accuracy:", accuracy_NB)
print("\nConfusion Matrix:")
print(conf_matrix_NB)
print("\nClassification Report:")
print(classification_rep_NB)
```

```python
# Evaluate the model Decision Tree Classifier 
accuracy_DTC = accuracy_score(y_test, y_pred2)
conf_matrix_DTC = confusion_matrix(y_test, y_pred2)
classification_rep_DTC = classification_report(y_test, y_pred2)

# Print the evaluation metrics
print("Accuracy:", accuracy_DTC)
print("\nConfusion Matrix:")
print(conf_matrix_DTC)
print("\nClassification Report:")
print(classification_rep_DTC)
```

```python
# Evaluate the model Logistic Regression 
accuracy_LR = accuracy_score(y_test, y_pred3)
conf_matrix_LR = confusion_matrix(y_test, y_pred3)
classification_rep_LR = classification_report(y_test, y_pred3)

# Print the evaluation metrics
print("Accuracy:", accuracy_LR)
print("\nConfusion Matrix:")
print(conf_matrix_LR)
print("\nClassification Report:")
print(classification_rep_LR)
```
